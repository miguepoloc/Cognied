## **Encuesta de Ansiedad**

Para esta encuesta se utilizaron los Baremos De Interpretación – Zung Self-Rating Anxiety Scale-15 (Zung, 1971 Adaptada Por De Ávila, León-Valle Et Al., 2021).

- 15-30 = **Ansiedad baja** = 1
- 31-45 = **Ansiedad media** = 2
- 46-60 = **Ansiedad alta** = 3


```python
listadic = []
encuesta = 3
for usuario in usuarios_id:
    dic_test = {}
    df_filtrado = usuario.loc[(df['id_usuario_encuesta__id_encuesta__id_encuesta'] == encuesta) & (
        df['id_usuario_encuesta__id_usuario__id'] == usuario)]
    id_encuesta = df_filtrado["id_usuario_encuesta__id_usuario_encuesta"].unique(
    )
    pretest = min(id_encuesta)
    posttest = max(id_encuesta)
    for tipo_test in id_encuesta:
        df_encuesta = df.loc[(
            df['id_usuario_encuesta__id_usuario_encuesta'] == tipo_test)]
        dic_test["id_usuario"] = df_encuesta['id_usuario_encuesta__id_usuario__id'].unique()[
            0]
        tipo = ""
        if tipo_test == pretest:
            tipo = "pre"
        elif tipo_test == posttest:
            tipo = "post"

        ansiedad = df_encuesta["id_pregunta_respuesta__id_respuesta__valor"].sum(
        )
        for i in range(0, len(df_encuesta)):
            dic_test[tipo + "_ansiedad" +
                     str(i)] = df_encuesta.iloc[i]["id_pregunta_respuesta__id_respuesta__valor"]

        dic_test[tipo + "_ansiedad_total"] = ansiedad
        if ansiedad in range(15, 31):
            dic_test[tipo + "_ansiedad_cat"] = 1
            dic_test[tipo + "_ansiedad_text"] = "Ansiedad baja"
        elif ansiedad in range(31, 45):
            dic_test[tipo + "_ansiedad_cat"] = 2
            dic_test[tipo + "_ansiedad_text"] = "Ansiedad media"
        elif ansiedad in range(45, 60):
            dic_test[tipo + "_ansiedad_cat"] = 3
            dic_test[tipo + "_ansiedad_text"] = "Ansiedad alta"

    listadic.append(dic_test)
resultado = pd.DataFrame(listadic)
resultado
```

- _Factor II_. **Claridad de sentimientos** Se suman las preguntas de la 9 a la 16
- _Factor III_. **Reparación del estado emocional** Se suman las preguntas de la 17 a la 24

| Hombres                         | Mujeres                         | Categorización |
